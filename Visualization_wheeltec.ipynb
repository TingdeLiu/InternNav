### Gif

##### 真实世界

```python
import imageio.v2 as imageio
from pathlib import Path

# 输入与输出路径
img_dir = Path("test_data/20260127_194712")
pattern = "vis_*.png"
out_gif =  "test_data/GIF/vis_20260127_194712.gif"

# 读取并按文件名排序
frames = []
for p in sorted(img_dir.glob(pattern)):
    frames.append(imageio.imread(p))

# 保存 GIF，duration 控制帧间隔（秒），或使用 fps 控制帧率
imageio.mimsave(out_gif, frames, fps=1)  # 1s per frame
print(f"GIF saved to: {out_gif}")
```

##### 仿真环境

```python
import imageio.v2 as imageio
from pathlib import Path

# 输入与输出路径
img_dir = Path("/data3/ltd/InternNav/logs/test_n1/video/7192_1808/frames")
pattern = "00*.png"
out_gif =  "test_data/GIF/7192_1808.gif"

# 读取并按文件名排序
frames = []
all_paths = sorted(img_dir.glob(pattern))
for p in all_paths[::3]:
    frames.append(imageio.imread(p))

# 保存 GIF，duration 控制帧间隔（秒），或使用 fps 控制帧率
imageio.mimsave(out_gif, frames, fps=12)  # 1s per frame
print(f"GIF saved to: {out_gif}")
size_mb = Path(out_gif).stat().st_size / (1024 * 1024)
print(f"成功保存！帧数: {len(frames)} | 大小: {size_mb:.2f} MB")
```
